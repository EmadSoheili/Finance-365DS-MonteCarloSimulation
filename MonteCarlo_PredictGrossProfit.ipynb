```python
# Goal: Predict firm's future gross profit
```

```python
# Import dependencies

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define variables and amounts
# rev_m: expected revenue mean for next year (in a million $)
# rev_stdev: expected revenue std for next year (in a million $)

rev_m = 170
rev_stdev = 20
iteration = 1000
```

```python
# create 1000 simulation of next year revenue

rev = np.random.normal(rev_m, rev_stdev, iteration)
rev
```

```python
# Plot revenue

plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
# Assumption: Based on previous experience, COGS is approximately 60% of revenue with 10% std

COGS = - (rev * np.random.normal(0.6,0.1))
```

```python
# Plot COGS
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
# Print outputs

print(f'''COGS mean: {round(COGS.mean(),1)} m$
COGS standard deviation: {round(COGS.std(),1)} m$''')
```

```python
# Gross profit

gross_profit = rev + COGS
gross_profit
```

```python
# Plot Gross profit

plt.figure(figsize=(15,6))
plt.plot(gross_profit)
plt.show()
```

```python
plt.show()
# Print outputs

print(f'''Gross profit mean: {round(gross_profit.mean(), 1)} m$
Gross profit standard deviation: {round(gross_profit.std(), 1)} m$
Gross profit minimum: {round(gross_profit.min(), 1)} m$
Gross profit maximum: {round(gross_profit.max(), 1)} m$''')
```

```python
# Plot a Histogram for Gross profit

plt.figure(figsize=(10,6))
plt.hist(gross_profit, bins=[i for i in range(40, 130, 5)])
plt.show()
```
